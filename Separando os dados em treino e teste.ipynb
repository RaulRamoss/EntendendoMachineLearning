# Separando os dados em treino e teste
- Qual o motivo de fazermos isso?
    - Nós vamos usar os dados do passado para criar um modelo e então vamos usar esse modelo para fazer previsões sobre o futuro

- O que nós queremos é avaliar qual vai ser a qualidade do modelo para esses dados do futuro

- Só que nós não temos esses dados
- Então uma forma de avaliar o modelo é usando os dados históricos e para isso vamos separar esses dados em treino e teste

- Além disso, também podemos separar os nossos dados em treino, validação e teste onde:
    - **Treino**: vamos usar para **ajustar** o modelo
    - **Validação**: usar para **otimizar** o modelo (ajustar parâmetros, testar diferentes hiperparâmetros, realizar melhorias, etc)
    - **Teste**: **avaliar** meu modelo final **já ajustado**

### Para fazer isso, podemos usar o train_test_split do Scikit-Learn
- https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

```python
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
base = pd.read_excel("BaseInadimplencia.xlsx")
base.head()
```

```python
X = base[['SaldoConta','SaldoInvestimento']]
y = base['Situacao']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```
